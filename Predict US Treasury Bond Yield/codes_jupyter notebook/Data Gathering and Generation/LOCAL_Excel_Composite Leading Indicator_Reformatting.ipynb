```python
### print original format
##### the desirable format is:  country name as column name, each column contains only the 'Value' variable ,index is TIME column
##### (see the final result in the last cell)
dt=pd.read_csv(r'C:\Users\Admin\Downloads\CL.csv')
print(dt)
```

```python
##get all country names
loc=pd.unique(dt['LOCATION'])
```

```python
d={}
#put different countries' data into a dictionary
for i in loc:
    df1 = dt[dt['LOCATION'] ==i]
    d[i]=df1
            
```

```python
###drop columns named ['INDICATOR','SUBJECT','MEASURE','FREQUENCY','Flag Codes','LOCATION']
###reset dataframe index to datetime formate using the 'TIME' column
for i in d:
    d[i]=d[i].set_index('TIME')
    d[i]=d[i].drop(columns=['INDICATOR','SUBJECT','MEASURE','FREQUENCY','Flag Codes','LOCATION'])
```

```python
### rename column from 'value' to 'country_name CLI'
for i in d:
    d[i]=d[i].rename(columns={'Value': "{}CLI".format(i)})
```

```python
for i in d:
    psave(d[i],i)
```
